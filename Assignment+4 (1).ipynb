```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler

def salary_predictions():

    df = pd.DataFrame(index=G.nodes())

    df['Department'] = pd.Series(nx.get_node_attributes(G, 'Department'))
    df['ManagementSalary'] = pd.Series(nx.get_node_attributes(G, 'ManagementSalary'))
    df['clustering'] = pd.Series(nx.clustering(G))
    df['degree'] = pd.Series(G.degree())
    df['degree_centrality'] = pd.Series(nx.degree_centrality(G))
    df['closeness'] = pd.Series(nx.closeness_centrality(G))
    df['betweeness'] = pd.Series(nx.betweenness_centrality(G))
    df['eigen'] = pd.Series(nx.eigenvector_centrality(G))
    df['pr'] = pd.Series(nx.pagerank(G))
    df[df.ManagementSalary.isnull()]
    
    
    features = ['clustering', 'degree', 'degree_centrality', 'closeness', 'betweeness', 'eigen', 'pr']

    scaler=MinMaxScaler()

    training = df[df.ManagementSalary.notnull()].copy()
    training_features = training[features]
    training_scaled = scaler.fit_transform(training_features)
    training_labels = training['ManagementSalary']

    testing = df[df.ManagementSalary.isnull()].copy()
    testing_features = testing[features]
    testing_scaled = scaler.fit_transform(testing_features)
    testing_labels = testing['ManagementSalary']

    rfc = RandomForestClassifier(n_estimators=500)
    rfc.fit(training_scaled, training_labels)

    answer = pd.Series(rfc.predict_proba(testing_scaled)[:, 1], testing_features.index)
    
    
    return answer
#salary_predictions()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler

def new_connections_predictions():
    
    # set the community attribute of each node using department.
    # This is required for community metrics such as Soundarajan-Hopcroft
    # and resource allocation index

    for node in G.nodes():
        G.node[node]['community'] = G.node[node]['Department']

    # create an index of all possible connections.  This is the same row size
    # as future_prediction
    big_index = []
    for i in list(nx.preferential_attachment(G)):
        big_index.append((i[0],i[1]))

    df = pd.DataFrame(index=big_index)

    pref_attach = [i[2] for i in list(nx.preferential_attachment(G))]
    df['preferential_attachment'] = pref_attach

    jaccard = [i[2] for i in list(nx.jaccard_coefficient(G))]
    df['jaccard_coefficient'] = jaccard

    # Community metrics
    # Common neighgor Soundarajan-Hopcroft score
    soundgarden = [i[2] for i in list(nx.cn_soundarajan_hopcroft(G))]
    df['soundgarden'] = soundgarden

    # Soundarajan-Hopcroft resource allocation index
    ra_index = [i[2] for i in list(nx.ra_index_soundarajan_hopcroft(G))]
    df['ra_index'] = ra_index

    merged_df = pd.merge(df, future_connections, right_index=True, left_index=True)

    features = ['preferential_attachment', 'jaccard_coefficient', 'soundgarden', 'ra_index']

    features = ['preferential_attachment', 'jaccard_coefficient', 'soundgarden', 'ra_index']
    scaler = MinMaxScaler()

    training = merged_df[merged_df['Future Connection'].notnull()].copy()
    training_features = training[features]
    training_scaled = scaler.fit_transform(training_features)
    training_labels = training['Future Connection']

    testing = merged_df[merged_df['Future Connection'].isnull()].copy()
    testing_features = testing[features]
    testing_scaled = scaler.fit_transform(testing_features)
    testing_labels = testing['Future Connection']

    rfc = RandomForestClassifier(n_estimators=100)
    rfc.fit(training_scaled, training_labels)

    answer = pd.Series(rfc.predict_proba(testing_scaled)[:, 1], testing_features.index)    
    
    return answer
#new_connections_predictions()
```
